```python
import matplotlib.pyplot as plt
import random

from src.face_segmentation_dataset import FacialSegmentationDataset

DATA_FOLDER = 'V2'  # data in V2 folder
```

```python
trainset = FacialSegmentationDataset(root_dir=DATA_FOLDER, train=True)
testset = FacialSegmentationDataset(root_dir=DATA_FOLDER, train=False)
```

```python
print(f"train set size = {len(trainset)}")
print(f"test set size = {len(testset)}")
```

```python
to_plot = 3
random_samples_indices = random.sample(range(len(trainset)), to_plot)

fig, axes = plt.subplots(nrows=to_plot, ncols=2, figsize=(8,10))

for i, sample_idx in enumerate(random_samples_indices):
    img, mask = trainset[sample_idx] # change to trainset[i] for non-random output
    axes[i,0].axis("off")
    axes[i,0].set_title("sample")
    axes[i,0].imshow(img)
    
    axes[i,1].axis("off")
    axes[i,1].set_title("mask")
    axes[i,1].imshow(mask)

fig.suptitle(f"{to_plot} Random samples from the train set", fontsize=20)     
fig.tight_layout()
```
